```
for model in models:
  model.compile(optimizer="Adam", loss="mse", metrics=[tf.keras.metrics.RootMeanSquaredError()])
```

```
def lrscheduler(epoch):
  
  if epoch <= 200:
    lrate = (10 ** -5) * (epoch / 100) 
  else:
    initial_lrate = (10 ** -5)
    k = 0.001
    lrate = initial_lrate * math.exp(-k * (epoch - 100))
  
  return lrate
```

```
callback = tf.keras.callbacks.LearningRateScheduler(lrscheduler)
```

```
c=1
for model in models:
  print('-'*15,"MODEL #",c,'-'*15)
  c+=1
  history = model.fit(x,y,epochs=800,verbose=0,callbacks=[callback])

  rmse = history.history['root_mean_squared_error']
  loss = history.history['loss']
  epochs = [i+1 for i in range(500)]

  p = figure(plot_width=800, plot_height=250)
  p.title.text = 'Click on legend entries to mute the corresponding lines'

  for data, name, color in zip([rmse, loss], ['RMSE', 'loss'], Spectral4):
    p.line(epochs ,data, line_width=2, color=color, alpha=0.8,
           muted_color=color, muted_alpha=0.2, legend_label=name)

  p.legend.location = "top_left"
  p.legend.click_policy="mute"

  show(p)
```

```
import math
from sklearn.metrics import mean_squared_error, mean_absolute_error
x_test = []
y_test = []
for i in range(len(apple_test)-5):
  x_test.append(apple_test[i:i+5])
  y_test.append(apple_test[i+5])

x_test = np.array(x_test).reshape(-1, 5, 1)
y_test = np.array(y_test).reshape(-1, 1)
c=1
pred_list = []

for model in models:
  predictions = model.predict(x_test)
  predictions = np.array(predictions).reshape(-1,1)
  pred_list.append(predictions)
  rmse = math.sqrt(mean_squared_error(predictions, y_test))
  print("MODEL #",c," rmse",rmse)
  c+=1
```
